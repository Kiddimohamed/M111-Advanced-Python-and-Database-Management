## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('../data/company_sales_data.csv')


x = data['month_number']
y = data['total_profit']
plt.figure()
plt.plot(x,y)
plt.xlabel("Month Number")
plt.ylabel("Profit in dollar")
plt.title("company profit for month")
plt.show()
```

<img src="../data/Profit_data.png" alt="Profit_data" style="width: 700px;"/>

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
plt.figure(figsize=(10, 6), dpi=80)
date=data['month_number']
plt.plot([],[],color="red" ,label='face cream sales data',marker = 'o')

plt.plot([],[],color="orange" ,label='face cream sales data',marker = 'o')

plt.plot([],[],color="green",label='toothpaste sales data',marker = 'o' )

plt.plot([],[],color="black",label='bathingsoap sales data',marker = 'o' )

plt.plot([],[],color="yellow" ,label='shampoo sales data',marker = 'o')

plt.plot([],[],color="blue" ,label='moisturizer sales data',marker = 'o')
plt.stackplot(date,data["facecream"],data["facewash"], data["toothpaste"], data["bathingsoap"],
              data["shampoo"],data["moisturizer"], baseline ='zero',
              colors =['red', 'orange',
                       'green','black','yellow','blue'])
plt.legend(loc='upper left')
plt.xlabel("Month number")
plt.ylabel("salles units in number")
plt.title("All product sales data using stack plot")


plt.show()
```
