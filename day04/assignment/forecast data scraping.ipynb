```python
!pip install seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup
import requests
```

```python
url1 = "https://weather.com/fr-MA/temps/10jours/l/f37580e42d804f6d54467e56dd9f296e15fe9a25db53ddee63601416ecc06111"
html = urlopen(url1)

soup = BeautifulSoup(html)
type(soup)
```

```python
container = soup.find('div', id="WxuDailyCard-main-a43097e1-49d7-4df7-9d1a-334b29628263")
days = container.find_all('h3')
get_days = []
for day in days:
    temp = day.find('span')
    if  temp != None:
        if temp.get_text() not in get_days:
            get_days.append(temp.get_text())
get_days
```

```python
container = soup.find('div', id="WxuDailyCard-main-a43097e1-49d7-4df7-9d1a-334b29628263")
container
summaries = container.find_all('summary')

def get_day_info(summary):
    data = []
    
    for divs in summary.find('div').find_all('div'):
        spans = divs.find_all('span')
        for span in spans:
            data.append(span.get_text())

        break
        
    return data
data = [get_day_info(summaries[day]) for day in range(15)]
data
```

```python
columns = ['day', 'temperature_max', 'temperature_min', 'state', 
           'percentage_value', 'wind_speed']
df = pd.DataFrame(data=data)
df.drop(1, axis=1, inplace=True)
df = pd.concat([pd.DataFrame(data=get_days), df], axis=1)
df.columns = columns
df
```
